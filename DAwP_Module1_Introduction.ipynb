```python
# Importing and Exporting Data in Python

# import pandas library
import pandas as pd
import numpy as np
```

```python
path = "sample_data/auto.csv"

df = pd.read_csv(path)
```

```python
df.head()
```

```python
# Question #1: Check the bottom 10 row of dta fram "df".
df.tail(10)
```

```python
df.describe()
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
# replace headers
df.columns = headers
df.head(10)
```

```python
# replace "?" with NaN so dropna() can remove missing values
df1=df.replace('?',np.NaN)
```

```python
df1.head()
```

```python
df = df1.dropna(subset=['price'], axis=0)
```

```python
df.head(20)
```

```python
# Question 3: Apply the method to ".describe()" to the columns 'length' and 'compression-ratio'.

df[['length', 'compression-ratio']].describe()
```
